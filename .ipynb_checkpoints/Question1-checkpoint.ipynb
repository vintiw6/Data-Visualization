```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Housing.csv")
```

```python
df.shape
```

```python
df.head(15)
```

```python
binary_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
df[binary_columns] = df[binary_columns].map(lambda x: 1 if x == 'yes' else 0)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="area", y="price", hue="furnishingstatus", size="bedrooms", palette="viridis")
plt.title("House Price vs Area with Furnishing and Bedrooms")
plt.xlabel("Area (sqft)")
plt.ylabel("Price")
plt.legend(title="Furnishing & Bedrooms")
plt.show()
```

```python
# Distribution of price based on furnishing status
plt.figure(figsize=(8, 5))
sns.boxplot(x="furnishingstatus", y="price", hue="furnishingstatus", data=df, palette="coolwarm", legend=False)
plt.title("Price Distribution by Furnishing Status")
plt.xlabel("Furnishing Status")
plt.ylabel("Price")
plt.show()
```

```python
# Heatmap of correlations (excluding non-numeric columns)
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
