```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Rainfall.csv")
```

```python
numeric_cols = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', 'ANNUAL']
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Rainfall Correlation Heatmap - Coolwarm Scale")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="viridis", fmt=".2f")
plt.title("Rainfall Correlation Heatmap - Viridis Scale")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="JUN", y="ANNUAL", hue="JUL", palette="plasma")
plt.title("June vs Annual Rainfall with July as Hue - Plasma Scale")
plt.xlabel("June Rainfall (mm)")
plt.ylabel("Annual Rainfall (mm)")
plt.show()
```
