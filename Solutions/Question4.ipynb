```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew
```

```python
# Generates skewed data (right-skewed)
np.random.seed(42)
data = np.random.exponential(scale=2, size=1000)  # Exponential distribution
```

```python
# Check the skewness before transformation
original_skewness = skew(data)
print(f"Original Skewness: {original_skewness:.2f}")
```

```python
# Plot original distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True)
plt.title("Original Skewed Data")
```

```python
# Apply Log Transformation
log_data = np.log1p(data)  # log(1 + x) to handle zeros
log_skewness = skew(log_data)
```

```python
# Apply Square Root Transformation
sqrt_data = np.sqrt(data)
sqrt_skewness = skew(sqrt_data)
```

```python
# Plot transformed distributions
plt.subplot(1, 2, 2)
sns.histplot(log_data, bins=30, kde=True, color="green")
plt.title("Log Transformed Data")
```

```python
plt.show()

# Print skewness after transformation
print(f"Log Transformed Skewness: {log_skewness:.2f}")
print(f"Square Root Transformed Skewness: {sqrt_skewness:.2f}")
```
