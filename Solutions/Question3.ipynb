```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"],
    "Temperature": [30, 35, 28, 33, 40],
    "Humidity": [60, 50, 55, 65, 30],
    "Wind Speed": [12, 8, 15, 10, 7]
}
```

```python
df_sample = pd.DataFrame(data)
df_sample.to_csv("sample_weather.csv", index=False)
```

```python
df = pd.read_csv("sample_weather.csv")
```

```python
numeric_cols = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', 'ANNUAL']
```

```python
# Bar plots for variables in sample dataset
plt.figure(figsize=(10, 6))
sns.barplot(x="City", y="Temperature", hue="City", data=df_sample, palette="coolwarm",  legend=False)
plt.title("City-wise Temperature Distribution")
plt.xlabel("City")
plt.ylabel("Temperature")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="City", y="Wind Speed", hue="City", data=df_sample, palette="coolwarm", legend=False)
plt.title("City-wise Wind Speed Distribution")
plt.xlabel("City")
plt.ylabel("Wind Speed")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="City", y="Humidity", hue ="City", data=df_sample, palette="magma", legend = False)
plt.title("City-wise Humidity Distribution")
plt.xlabel("City")
plt.ylabel("Humidity")
plt.xticks(rotation=45)
plt.show()
```
